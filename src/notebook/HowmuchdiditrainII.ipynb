# Exploratory Analysis

Importing basic libraries :

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Importing training data

```python
train = pd.read_csv("../../data/train.csv")
```

Preview of this data

```python
train.head(20).transpose()
```

Overall statistics of all columns.
Can some columns be dropped because they don't add any value?

```python
train.describe().transpose()
```

As a naive start, I would get rid of all the records which are having even field as NA. This would greatly reduce the overall data size and would also provide a good baseline number.
Also easier to handle.

After dropping NA columns, Printing new summary

```python
train = train.dropna();
```

```python
train.describe().transpose()
```

How many data points have rainfall above 300? Anything above that seems odd, given the percentiles. Will get rid of those.

```python
train.drop(train[train['Expected'] > 500].index).describe().transpose()
```

With these remaining data points we will continue with our study.

```python
train = train.drop(train[train['Expected'] > 500].index)
```

## Analyze expected rainfall

```python
plt.figure()
plt.hist(train['Expected'].as_matrix(), bins = 1000)

#plt.hist(log)

plt.title("Expected Rainfall histogram")
plt.xlabel("Expected Rain")
plt.ylabel("Frequency")
plt.show()
```

Export this data out for learning

```python
plt.figure()
plt.hist(train['RhoHV'].as_matrix(), bins=50)
plt.title("Expected RhoHV reading histogram")
plt.xlabel("RhoHV")
plt.ylabel("Frequency")
plt.show()
```

```python
train.to_csv("../../data/train_processed.csv",index=False)
```
